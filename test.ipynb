## Kmeans

```python
X = np.random.randn(500,2)
kmeans = KMeans(4)
kmeans.train(X)
for i in range(len(np.unique(kmeans.predict()))):
    points = X[kmeans.predict()==i,:]
    plt.scatter(points[:,0], points[:,1])
plt.show()
```
